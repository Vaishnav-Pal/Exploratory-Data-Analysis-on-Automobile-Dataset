## Missing Data

### Fill the missing data in the normalized-losses, price, horsepower, and peak-rpm columns with their respective column means.

```python
data=df['normalized-losses'].loc[~df['normalized-losses'].isna()]
mean=data.astype(int).mean()
df['normalized-losses']=df['normalized-losses'].replace(np.nan,mean).astype(int)
df['normalized-losses'].head()
```

```python
data=df['price'].loc[~df['price'].isna()]
mean=data.astype(int).mean()
df['price']=df['price'].replace(np.nan,mean)
df['price'].head()
```

```python
df['price'].unique()
```

```python
data = df['horsepower'].loc[~df['horsepower'].isna()]
mean = data.astype(int).mean()
df['horsepower'] = df['horsepower'].replace(np.nan,mean).astype(int)
df['horsepower'].head()
```

```python
data = df['peak-rpm'].loc[~df['peak-rpm'].isna()]
mean = data.astype(int).mean()
df['peak-rpm'] = df['peak-rpm'].replace(np.nan,mean).astype(int)
df['peak-rpm'].head()
```

# Show a distribution chart for the price

#### The distribution in question is "right-skewed," meaning its tail extends towards the right. Positive skewness indicates that the right tail is longer or fatter compared to the left tail.
#### The effects of skewness on the data are as follows:
 - The mean is greater than the mode.
 - The median is greater than the mode.
 - The mean is greater than the median.
#### In this case, most car prices range between $5,000 and $20,000, with only a few cars being significantly more expensive.
#### Kurtosis measures the heaviness of the tails relative to the rest of the distribution. A positive kurtosis value indicates that the dataset has heavier tails, meaning more extreme values are present compared to a normal distribution.

```python
sns.displot(df_copy['price'],kde=True,color='red')
print('This distribution has skew =', df['price'].skew())
print('This distribution has kurtosis =', df['price'].kurt())
```

# Normalized Losses Across Body Styles and Door Counts 


#### **Findings**:The normalized loss, representing the average loss payment per insured vehicle, is influenced by various car features, including body style and the number of doors. The distribution of normalized losses varies by body style, with two-door cars generally experiencing higher losses compared to four-door cars.

```python
final_find=pd.pivot_table(df_copy,index=['body-style','num-of-doors'], values='normalized-losses')
final_find
```

```python
final_find.plot(kind='bar',color='orange',
title='Normalized losses based on body style and num of doors',xlabel='Body style and No. of doors',ylabel='Normalized losses',fontsize=15)
```
